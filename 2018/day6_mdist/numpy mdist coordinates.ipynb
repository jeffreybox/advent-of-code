```python
import numpy as np
```

```python
coords = np.loadtxt('coordinates.txt', delimiter=', ')
print(coords[0])
print(coords[0,0])
print(coords)
```

```python
# meshgrid returns coordinate matrices from coordinate vectors.
# notes on meshgrid: https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.meshgrid.html
xvalues = np.arange(coords[:,0].max())
yvalues = np.arange(coords[:,1].max())
xx, yy = np.meshgrid(xvalues, yvalues)
print(xx)
print(yy)
```

```python
layers = []
for coord in coords:
    mdists = np.abs(xx - coord[0]) + np.abs(yy - coord[1])
    layers.append(mdists)
    
dist_arr = np.array(layers)
tot_dists = dist_arr.sum(axis=0)
```

```python
print(dist_arr.max())
print(dist_arr) 
```

```python
print(tot_dists.max())
print(tot_dists) 
```
