## Part II
What is the first frequency your device reaches twice?

```python
# extending the list
new_freq_list = []
i = 0

while i < 1000:
    new_freq_list.extend(freq_list)
    i = i +1
    
print(len(freq_list))
print(len(new_freq_list))
```

```python
cumfreq_list = []
prev_freq = 0

for freq in new_freq_list:
    x = freq + prev_freq
    prev_freq = x
    cumfreq_list.append(x)
    
print(len(cumfreq_list))
```

```python
def firstDuplicate(a):
    set_ = set()
    for item in a:
        if item in set_:
            return item
        set_.add(item)
    return None
```

```python
answer = firstDuplicate(cumfreq_list)
print(answer)
```
