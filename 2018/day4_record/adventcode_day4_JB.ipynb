```python
import re
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('records.csv', header=None)
df.head()
```

```python
with open('records.txt') as f:
    inp = []
    for r in f.readlines():
        # Matches any character which is not a decimal digit --> [^0-9].
        r = re.split('[^0-9]+', r[1:].strip())
        inp.append([d for d in r])
        
print(inp)
```
