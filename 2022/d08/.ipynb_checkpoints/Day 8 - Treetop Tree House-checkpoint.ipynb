### DATA

```python
path = 'data/aoc2022day8.txt'
```

```python
with open(path) as f:
    treemap = [line.rstrip() for line in f]

print(f'length = {len(treemap)}')
print(f'test output... {treemap[:3]}')
```

```python
for tree in treemap:
    print(tree)
```

### PART 1

```python
# first count all trees on the exterior

topcnt = len(treemap[0])
print(f'top & bottom = {topcnt}')
bottomcnt = len(treemap[0])
leftcnt = len(treemap)-2 # accounting for edges
print(f'left & right subtracting edges = {leftcnt}')
rightcnt = leftcnt
edgecnt = topcnt+bottomcnt+leftcnt+rightcnt
print(f'total trees around edges = {edgecnt}')
```

```python
# set up some variables 

x = 1
y = 1
innerforest = len(treemap[1])
print(innerforest)
treemap[x][y]
```

```python
# create the view from the left 

def create_left(x,y):
    left = []
    for i in range(y):
        left.append(int(treemap[x][i]))
    left = list(reversed(left))
    return left
```

```python
# create the view from the right 

def create_right(x,y):
    right = []
    for i in range(y+1,innerforest):
        right.append(int(treemap[x][i]))
    #right = list(reversed(right))
    return right
```

```python
# create the view from the top

def create_top(x,y):
    top = []
    for i in range(x):
        top.append(int(treemap[i][y]))
    top = list(reversed(top))
    return top
```

```python
# create the view from the bottom

def create_bottom(x,y):
    bottom = []
    for i in range(x+1,innerforest):
        bottom.append(int(treemap[i][y]))
    #bottom = list(reversed(bottom))
    return bottom
```

```python
# iterate through the trees
viewcount = 0

for j in range(1,innerforest-1):
    for k in range(1,innerforest-1):
        tree = int(treemap[j][k])
        if all(l < tree for l in create_left(j,k)) \
        or all(r < tree for r in create_right(j,k)) \
        or all(t < tree for t in create_top(j,k)) \
        or all(b < tree for b in create_bottom(j,k)):
            viewcount += 1

print(viewcount+edgecnt)
```

### PART 2

```python
def scorer(ls, tree):
    #print(ls)
    score = 1
    if tree <= ls[0]:
        return score
    else:
        for i, j in enumerate(ls[:-1]):
            if tree > j:
                score +=1
            else:
                break
    return score
```

```python
# create scenic scores
scorelist = []

for j in range(1,innerforest-1):
    for k in range(1,innerforest-1):
        tree = int(treemap[j][k])
        lscore = scorer(create_left(j,k), tree)
        rscore = scorer(create_right(j,k), tree)
        tscore = scorer(create_top(j,k), tree)
        bscore = scorer(create_bottom(j,k), tree)
        totalscore = lscore * rscore * tscore * bscore
        scorelist.append(totalscore)
```

```python
max(scorelist)
```
